```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
#simple example 
x = np.arange(1,10)
y = np.arange(11,20)
```

```python
#scatter plt 
plt.scatter(x,y,c='r')      #imp hai ye 
plt.plot(x,y,'g--')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('2d arrray')
```

```python
np.pi
```

```python
x= np.arange(0,8*np.pi,0.2)
y= np.sin(x)
plt.title('smit')

#plot using matplot
plt.plot(x,y)
plt.show()
```

```python
x= np.arange(0,8*np.pi,0.2)
y= np.sin(x)
yc= np.cos(x)

#plot using matplot
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('smit')

plt.subplot(2,1,2)
plt.plot(x,yc)
plt.title('smittt')
plt.show()
```
