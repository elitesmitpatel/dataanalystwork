```python
##closure


def fun1():
    
    a='i m a good coder'
    
    def fun2():
        print ('ya u r')
        
        print(a)
        
        print ('u r also a good trader')
    return fun2()    
```

```python
fun1()
```

```python
##closure

def fun1(fun):
    
    def fun2():
        print ('ya u r')
        fun("i m a good coder ")
        print ('u r also a good trader')
    return fun2()    
```

```python
fun1(print)
```

```python
##closure & initial decoratores

def fun1(fun):
    
    def fun2():
        print ('ya u r')
        print(fun([1,2,3,4,5,6,7]))                           #function
        print ('u r also a good trader')
    return fun2()    
```

```python
fun1(len)
```

```python
## decoratores

def fun1(fun):
    
    def fun2():
        print ('ya u r')
        fun()                          
        print ('u r also a good trader')
    return fun2()    
```

```python
def bharma():
    print('smit')
```

```python
fun1(bharma)
```
