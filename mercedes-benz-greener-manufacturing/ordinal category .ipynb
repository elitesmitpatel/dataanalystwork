```python
import pandas as pd
import datetime
```

```python
# create a variable with dates ,and from that extract the week days
# create a list of date with 20 days differnece from today and then transform into data frame 

df_base = datetime.datetime.today()
df_date_list =[df_base - datetime.timedelta(days=x) for x in range (0,20)]
df= pd.DataFrame(df_date_list)
df.columns =['Days']
df
```

```python
#extract days of week 
df['days_of_week'] = df['Days'].dt.day_name()                 #yaha   dt.weekday_name   use nahi hua 
df.head()
```

```python
#Enginer ctegorical varible by ordinal number replacement

weekday_map = {'Monday': 1,
              'Tuesday': 2,
              'Wednesday':3,
              'Thursday' : 4,
              'Friday' : 5,
              'Saturday' : 6,
              'Sunday' : 7 }

df['day_ordinal'] = df.days_of_week.map(weekday_map)
df.head(20)
```
